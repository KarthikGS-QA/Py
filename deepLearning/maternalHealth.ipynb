Let us define ANN (Artificial Neural Networks) to classify Maternal risk

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Load the Dataset
df=pd.read_csv("Maternal Health Risk Data Set.csv")
```

```python
print(df.head())
```

```python
df.head()
```

```python
#Let us check null values
df.isnull().sum()
```

```python
#Let us split features and target
X=df.drop("RiskLevel", axis=1)   #Inserting all features to X by dropping RiskLevel column
y=df['RiskLevel']                #Inserting RiskLevel column feature to y axis
```

```python
#Let us encode the target (converting string to interger of given data set of the target (risk, mid risk, high risk into 0,1,2))
encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
y
```

```python
#Let us split train and test data

Xtrain, Xtest, ytrain, ytest=train_test_split(X,y,test_size=0.2, random_state=43)
```

```python
#Features Scaling

scaler=StandardScaler()
Xtrain=scaler.fit_transform(Xtrain)
Xtest=scaler.transform(Xtest)
```

```python
Xtrain[0]
```

```python
Xtest[0]
```

```python
#Let us Build Model
model=Sequential()
model.add(Dense(16, activation="relu", input_shape=(6,)))           #6 features(6 columns)
model.add(Dense(8, activation="relu"))
model.add(Dense(3, activation="softmax"))       #Since it is multiclass output we using softmax activation type. For Binary class we use Sigmoid
```

```python
#Let us set compile option

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"])
```

```python
history=model.fit(
    Xtrain,
    ytrain,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

```python

print("Training accuracy: ", history.history['accuracy'][-1]*100)
print("Validation accuracy: ", history.history['val_accuracy'][-1]*100)
```

```python
test_data=np.array([[25,13,80,15,98,86]])
test_data=scaler.transform(test_data)
pred=model.predict(test_data)
print(f"Result of prediction ={pred[0]}")

index=np.argmax(pred)
print(f"Predicted index is: {index}")

#Test data and validaion data is different
#total data=1000, 800=train data, 200=test data. in 800 train data 640 is trained data and 160 is validation data
```

```python
ypred=model.predict(Xtest)
# ypred=np.argmax(ypred,axis=1)

ypred           #output will be in probability. 
```

```python
ypred=np.argmax(ypred,axis=1)       #Converting the ypred data from pobablity to decision

ypred
```
