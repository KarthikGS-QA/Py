```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plot
import os 
```

```python
#step 2: Define dataset paths

Train_Dir="./Tomato_Split/train"
Validation_Dir="./Tomato_Split/validation"
Test_Dir="./Tomato_Split/test"
```

```python
#Step 3: Set Image parameters

IMG_HEIGHT=224
IMG_WIDTH=224
BATCH_SIZE=32
EPOCHS=10
```

```python
#Step 4: Image Preprocessing and Augmentation

train_datagen=ImageDataGenerator(

    rescale=1./225,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True

)
validation_datagen=ImageDataGenerator(

    rescale=1./255

)
```

```python
#step 5: Load images from folders

train_data=train_datagen.flow_from_directory(

    Train_Dir,
    target_size=(IMG_HEIGHT,IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'

)

validation_data=validation_datagen.flow_from_directory(

    Validation_Dir,
    target_size=(IMG_HEIGHT,IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'

)

NUM_CLASS=train_data.num_classes
print("Number of classes : ", NUM_CLASS)
print("Class Lables : ",train_data.class_indices)
```

```python
#Step 6: Built CNN Model

model=Sequential()

model.add( Conv2D( 32,(3,3),activation='relu',input_shape=(IMG_HEIGHT,IMG_WIDTH,3)))
model.add(MaxPooling2D(2,2))

model.add( Conv2D( 64,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add( Conv2D( 128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())


model.add(Dense(128,activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(NUM_CLASS,activation='softmax'))
```

```python
model.summary()
```

```python
# Sterp 7: Compile the model

model.compile(

    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#step 8 Train the model

history=model.fit(

    train_data,
    validation_data=validation_data,
    epochs=EPOCHS
)
```

```python
# Step 10 : Save the model

model.save("tomato_plant_disease_cnn.h5")
print("Model saved")
```
