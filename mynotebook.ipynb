```python
from math import sqrt
def isPrime(n):
    for i in range(2,int(sqrt(n))):
        if n%i==0:
            return False
    
    return True
```

```python
for item in range(5,101):
    if isPrime(item):
        print(f" {item} is a prime number")
```

```python
# reverse a no

n=input("Enter a number")
n=int(n)
rev=0
sum=0

print(f"The given no is : {n}")

while(n!=0):

    rem=n%10
    sum=sum+rem
    rev=rev*10+rem
    # print(rem,end=",")
    n=n//10

print(f"\nThe reverse number of the given no is {rev}")
print(f"the sum of given no = {sum}")
```

# Inhetance
-> it is defined as process of creating a new class from existing class
-> exicting class is called parent class or base class
-> Newly created class is called sub class or derived class

Based on parent and derived class inheritance is classified into following 
1. Single      ( one parent one child)
2. Multilevel  ( a class derived from another derivied class)
3. Multiple    ( Multiple parent one child class)
4. Hybrid      ( Combination of above)


```python
# Single inheritance

class Data:
    x=10

    def __init__(self,y):
        self.y=y
    
    def display(self):
        print(f"X = {Data.x}, Y = {self.y}")
```

```python
class Sum(Data):
    def __init__(self, y):
        super().__init__(y)
    
    def findsum(self):
        sum=Data.x+self.y
        print(f"Sum of X, Y = {sum}")
```

```python
res=Sum(20)
res.display()
res.findsum()
```
