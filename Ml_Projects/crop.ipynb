```python
x=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
#Lets split our data into training and testing

from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
print(f"size of xtrain : {len(Xtrain)}, size of xtest : {len(Xtest)}")
```

```python
"""
Good classification algoritham for csv type of data
KNN
DecissionTree
Random Forest Classifier
SVM
Niave Bayes

"""
```

```python
#Import KNN class

from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier()

model.fit(Xtrain,Ytrain)                # fit trains the model


```

```python
import numpy as np

test_data= np.array([[100,40,35,27.56441788,54.41094079,6.9557873510000015,177.816092]])

pred=model.predict(test_data)

print(pred)
```

```python
# Evaluation of Model
from sklearn.metrics import accuracy_score

ypred=model.predict(Xtest)

score=accuracy_score(Ytest,ypred)

print(f"Accuracy score of KNN model is : {score*100}")
```

```python
yprl=list(ypred)
ytel=list(Ytest)
for i in range(0,len(yprl)):
    if(ypred[i]!=ytel[i]):
        print(f"{yprl[i]}\t{ytel[i]}")
```

```python
cropnames=df['label'].unique()
sorted(cropnames)
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(Ytest,ypred,labels=cropnames)
```

```python
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    cbar=True,
    xticklabels=cropnames,
    yticklabels=cropnames
)

```
